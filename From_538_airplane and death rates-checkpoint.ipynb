```python
def normalizeToTrillionAsk(rawNumber, ask, annualScalingFactor = 52, durationWindowInYears = 15):
    return (round(1e12*rawNumber/(annualScalingFactor*durationWindowInYears*ask)))
```

```python
fiveThirtyEightIncidentSeries85_99 = normalizeToTrillionAsk(fiveThirtyEightData['incidents_85_99'],fiveThirtyEightData['avail_seat_km_per_week'])
fiveThirtyEightFatalIncidentSeries85_99 = normalizeToTrillionAsk(fiveThirtyEightData['fatal_accidents_85_99'],fiveThirtyEightData['avail_seat_km_per_week'])
fiveThirtyEightFatalitySeries85_99 = normalizeToTrillionAsk(fiveThirtyEightData['fatalities_85_99'],fiveThirtyEightData['avail_seat_km_per_week'])
fiveThirtyEightIncidentSeries00_14 = normalizeToTrillionAsk(fiveThirtyEightData['incidents_00_14'],fiveThirtyEightData['avail_seat_km_per_week'])
fiveThirtyEightFatalIncidentSeries00_14 = normalizeToTrillionAsk(fiveThirtyEightData['fatal_accidents_00_14'],fiveThirtyEightData['avail_seat_km_per_week'])
fiveThirtyEightFatalitySeries00_14 = normalizeToTrillionAsk(fiveThirtyEightData['fatalities_00_14'],fiveThirtyEightData['avail_seat_km_per_week'])
```
